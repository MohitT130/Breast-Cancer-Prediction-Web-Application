```python
# here we are getting in terms of b and m. we will try to get in terms of 0 and 1 aur ye karne  kaa tarika hai lebel encoding which follos an arder
```

```python
#from sklearn.preprocessing import LabelEncoder
#le = LabelEncoder()
#here we are going to do pandas instead this we can use label encoding for vwry large dataset
```

```python
breast['diagnosis'] = breast['diagnosis'].map({"M":1, 'B':0})   
# here we are passing a dictionary to it, jisse labelling ki jaa sake in terms of binary(0 and 1)
```

```python
breast['diagnosis'].value_counts()
```

```python
breast.shape
```

```python
x = breast.drop('diagnosis', axis = 1)
y = breast['diagnosis']
```

```python
x.shape
```

```python
# Now splitting into training and tsting datasets
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# NOW WE ARE GOING FOR FEATURE SCALING
```

```python
x_train
```

```python
#the given set is not in standard form, we need to do scaling so that prediction can be inhanced
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(x_train)
```

```python
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
# we will be using logistic regression model here
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
```

```python
lg.fit(x_train,y_train)
y_pred = lg.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
